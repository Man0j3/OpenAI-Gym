```python
import gym #import OpenAI Gym
import numpy as np #Import numpy
```

```python
#This will run an instance of the CartPole-v0 environment for 1000 timesteps, rendering the environment at each step
env = gym.make('CartPole-v0')
observation=env.reset()
# observation  #[position of cart,velocity of cart,angle of pole,velocity of pole at its tip]
for _ in range(1000):
    env.render()
    env.step(env.action_space.sample()) # take a random action
env.close()
```

```python
#Making random movements
env = gym.make('CartPole-v0')
#play the game for ten times
for i_episode in range(10):
    observation = env.reset()
    print("observation at reset:",observation)
    #objective of this game is to keep the pore vertical as long as possible
    count=0
    done=False
    while not done:
        #it will take astronomical amount of time (avoid it for real reinforcement learning)
        env.render() #render the system to the screen (without this graphical view is not shown)
        count+=1
        print("observation:",observation)
        # here action is either moving left or right
        action = env.action_space.sample() #api will provide a random action here(0 or 1 here)
        observation, reward, done, info = env.step(action) # telling env to make one step or make one movement
        if done:  #come out of the game if done is true
            print("Episode finished after {} timesteps".format(count+1))
            break
env.close()
```

```python
#In the examples above, we’ve been sampling random actions from the environment’s action space. But what actually are those actions? Every environment comes with an action_space and an observation_space. These attributes are of type Space, and they describe the format of valid actions and observations
env = gym.make('CartPole-v0')
print(env.action_space)
#> Discrete(2)
print(env.observation_space)
```

```python
print(env.observation_space.high)
#> array([ 2.4       ,         inf,  0.20943951,         inf])
print(env.observation_space.low)
#> array([-2.4       ,        -inf, -0.20943951,        -inf])
```
